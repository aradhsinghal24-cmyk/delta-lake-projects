```python
%sql
-- 1. NOT NULL — column cannot be empty
-- 2. CHECK — custom condition must be true  
-- 3. Primary key / Foreign key (Unity Catalog only, informational)
```

```python
%sql
-- Cell 1: Create table with constraints
CREATE OR REPLACE TABLE delta_catalog.delta_demo.employees_constrained (
    id      INT NOT NULL,
    name    STRING NOT NULL,
    salary  INT,
    dept    STRING
)
USING DELTA;

-- Add a CHECK constraint
ALTER TABLE delta_catalog.delta_demo.employees_constrained
ADD CONSTRAINT salary_check CHECK (salary > 0);
```

```python
%sql
-- Cell 2: Valid insert — works fine
INSERT INTO delta_catalog.delta_demo.employees_constrained VALUES
(1, 'Aditi',  80000, 'Engineering'),
(2, 'Rahul',  90000, 'Marketing');
```

```python
%sql
-- Cell 3: Violate NOT NULL — should fail
INSERT INTO delta_catalog.delta_demo.employees_constrained VALUES
(3, NULL, 75000, 'Engineering');
```

```python
%sql
-- Cell 4: Violate CHECK — should fail
INSERT INTO delta_catalog.delta_demo.employees_constrained VALUES
(4, 'Priya', -5000, 'Engineering');
```

```python
%sql
-- Cell 5: See all constraints on the table
DESCRIBE DETAIL delta_catalog.delta_demo.employees_constrained;

-- Or
SHOW TBLPROPERTIES delta_catalog.delta_demo.employees_constrained;
```

```python
%sql
-- Cell 6: Drop a constraint
ALTER TABLE delta_catalog.delta_demo.employees_constrained
DROP CONSTRAINT salary_check;

-- Cleanup
DROP TABLE IF EXISTS delta_catalog.delta_demo.employees_constrained;
```
